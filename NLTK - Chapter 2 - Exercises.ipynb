◑ 17. Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
import nltk
from collections import Counter
from string import punctuation

def most_common_words(text):
    stopwords = set(nltk.corpus.stopwords.words('english')) # 0(1) lookups
    with_stp = Counter()
    without_stp  = Counter()
    for line in text:
        # update count off all words in the line that are in stopwrods
        with_stp.update(w.lower().rstrip(punctuation) for w in line if w.lower() in stopwords)
        # update count off all words in the line that are not in stopwords
        without_stp.update(w.lower().rstrip(punctuation)  for w in line if w  not in stopwords)
    # return a list with top ten most common words from each 
    return [x for x in with_stp.most_common(50)],[y for y in without_stp.most_common(50)]
    
most_common_words(brown.sents('cg05'))
```

◑ 18. Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
import nltk
from collections import Counter
from string import punctuation

def most_common_bigrams(text):
    stopwords = set(nltk.corpus.stopwords.words('english')) # 0(1) lookups
    with_stp = Counter()
    without_stp  = Counter()
    for line in text:
        big_line = nltk.bigrams(line)
        big_line = [' '.join(bigram) for bigram in big_line]
        # update count off all words in the line that are in stopwrods
        with_stp.update(w.lower().rstrip(punctuation) for w in big_line if any(stop in w.lower() for stop in stopwords))
        # update count off all words in the line that are not in stopwords
        without_stp.update(w.lower().rstrip(punctuation) for w in big_line if not any(stop in w.lower() for stop in stopwords))
    # return a list with top ten most common words from each 
    return [x for x in with_stp.most_common(50)],[y for y in without_stp.most_common(50)]
    
most_common_bigrams(brown.sents('cg10'))
```

◑ 20. Write a function word_freq() that takes a word and the name of a section of the Brown Corpus as arguments, and computes the frequency of the word in that section of the corpus.

```python
def word_freq(genre, word):
    cfd = nltk.ConditionalFreqDist((genre, word)
                                  for genre in genre
                                  for word in brown.words(categories=genre))

    cfd.tabulate(conditions=genre, samples=word)

word_freq(['news'], ['tomorrow'])
print()
word_freq(['news'], ['good'])
print()
word_freq(['romance'], ['like'])
```

★ Zipf's Law: Let f(w) be the frequency of a word w in free text. Suppose that all the words of a text are ranked according to their frequency, with the most frequent word first. Zipf's law states that the frequency of a word type is inversely proportional to its rank (i.e. f × r = k, for some constant k). For example, the 50th most common word type should occur three times as frequently as the 150th most common word type.

    a. Write a function to process a large text and plot word frequency against word rank using pylab.plot. Do you confirm Zipf's law? (Hint: it helps to use a logarithmic scale). What is going on at the extreme ends of the plotted line?

    b.Generate random text, e.g., using random.choice("abcdefg "), taking care to include the space character. You will need to import random first. Use the string concatenation operator to accumulate characters into a (very) long string. Then tokenize this string, and generate the Zipf plot as before, and compare the two plots. What do you make of Zipf's Law in the light of this?

```python
import nltk
from collections import Counter
from string import punctuation


def most_common_words(text):
    with_stp = Counter()
    without_stp  = Counter()
    # update count off all words in the line that are not in stopwords
    with_stp.update(word.lower().rstrip(punctuation) for word in text)
    # return a list with top ten most common words from each 
    return with_stp.most_common(10)
    
```

```python
from matplotlib import pyplot as plt
import numpy as np

genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
for genre in genres:
    common_words = most_common_words(brown.words(categories=genre))
    # save the names and their respective scores separately
    # reverse the tuples to go from most frequent to least frequent 
    people = [word[0] for word in common_words]
    score = [word[1] for word in common_words]
    x_pos = np.arange(len(people)) 

    plt.bar(x_pos, score,align='center')
    plt.xticks(x_pos, people) 
    plt.yscale('log')
    plt.ylabel('Popularity Score')
    plt.show()
```
